```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
print(df)
```

```python
y = df['Survived'].values.reshape(-1,1)
data = df.drop(["Name",'SibSp','Parch','Fare','Ticket','Cabin','Embarked','PassengerId','Survived'], axis=1)
print(data)
```

```python
print(data.isnull().sum())
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
print(data)
```

```python
data['Sex'] = data['Sex'].map({'male': 1 ,'female':2})

print(data)
```

```python
import math 
print(y.shape)
J = 0
m = data.shape[0]
# print(m)
alpha = 0.01

b = 1
w = np.ones(data.shape[1]).reshape(-1,1)
# print(w.shape)

for i in range(1000):

    z = np.dot(data,w)+b
    # print(z)
    def sigmoid(z):
        return (1/(1+np.exp(-z)))

    a = sigmoid(z)
    a = np.clip(a, 1e-15, 1-1e-15)
    # print(a.shape)
    J = -(1/m)*np.sum((y*np.log(a)+(1-y)*np.log(1-a)))
    da = np.sum((-y/a)+((1-y)/(1-a)))/m
    dz =a - y
    # print(dz.shape)
    dw = np.dot(data.T,dz)/m
    # print(dw.shape)
    db = np.sum(dz)/m
    b = b - alpha * db
    w = w - alpha * dw
    # print("loss for iteration", i, "is", J)

    # print("w for iteration", i, "is", w)
    # print("b for iteration", i, "is", b)
print("w = ",w ,'and b= ',b)    
```
